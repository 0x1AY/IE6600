### In Class Activity 1 (~ 15 min)

For the `penguins` dataset, do the following tasks:

- Visualize the ditribution of flipper length with a suitable number of bins. Have a density overlay on top of the histogram

- Visualize the distributions of the body masses for different species. The bars should not be stacked but overlain

- Make a chart with two panels for the two genders, showing KDE smoothings for the densities of each species

- Which island hosts the heaviest penguins? Is there a general trend?


```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

penguins = pd.read_csv('https://raw.githubusercontent.com/qurat-azim/instructionaldatasets/main/data/penguins.csv')
```

```python
#- Visualize the ditribution of flipper length with a suitable number of bins. Have a density overlay on top of the histogram

sns.displot(data=penguins, x="flipper_length_mm", bins=20, kde=True)
```

```python
#- Visualize the distributions of the body masses for different species. The bars should not be stacked but overlain
sns.displot(data=penguins, x="body_mass_g", hue='species', kind='hist', multiple='layer')
```

```python
# - Make a chart with two panels for the two genders, showing KDE smoothings for the densities of each species
sns.displot(data=penguins, x="body_mass_g", hue='species', kind='kde', col ='sex')
```

```python
### - Which island hosts the heaviest penguins? Is there a general trend?

penguins_body_mass = penguins.groupby(['island','species'])[['body_mass_g']].sum()
#heaviest island
heaviest_island = penguins_body_mass.groupby('island').sum().idxmax()[0]
print(f"The island that hosts the heaviest penguins is: {heaviest_island}")

# General trend
sns.boxplot(data=penguins, x='island', y='body_mass_g')
plt.title('Body Mass Distribution by Island')
plt.show()
trend_explanation = """
The boxplot shows the distribution of penguin body masses across different islands. 
From the plot, we can observe that Biscoe Island hosts the heaviest penguins, this is indicated by the higher median and upper quartile values compared to other islands.
This suggests that penguins on Biscoe Island generally have a higher body mass.
"""
print(trend_explanation)
```

### In Class Activity 2 (~10 min)

- Is the median body mass of a female `Gentoo` penguin bigger or smaller than the make counterpart? Use a boxplot to answer this

- Which island hosts the maximum number of penguins? Maximum number of male penguins? Maximum number of female penguins? Use only one visual to answer all three queries.

```python
# Filter the penguins dataset for Gentoo species
gentoo_penguins = penguins[penguins['species'] == 'Gentoo']

# Create a boxplot to compare the body mass of male and female Gentoo penguins
sns.boxplot(data=gentoo_penguins, x='sex', y='body_mass_g')
plt.title('Body Mass of Male and Female Gentoo Penguins')
plt.show()
```

```python
# Group by island and sex to get the count of penguins
penguin_counts = penguins.groupby(['island', 'sex']).size().reset_index(name='count')

# Plot the data
sns.barplot(data=penguin_counts, x='island', y='count', hue='sex')
plt.title('Penguin Counts by Island and Sex')
plt.show()
```

```python
#- Given the limited amount of data, can you tell which penguin type has is the most diverse in terms of flipper length?
penguinss = pd.read_csv('https://raw.githubusercontent.com/qurat-azim/instructionaldatasets/main/data/penguins.csv')

sns.boxplot(data=penguinss , x='species', y='body_mass_g')

#- Given the limited amount of data, can you tell which penguin type has is the most diverse in terms of flipper length?
# Calculate the standard deviation of flipper length for each species
flipper_length_std = penguins.groupby('species')['flipper_length_mm'].std()
print(flipper_length_std)
# Find the species with the highest standard deviation
most_diverse_species = flipper_length_std.idxmax()
diversity_value = flipper_length_std.max()

print(f"The most diverse penguin type in terms of flipper length is: {most_diverse_species} with a standard deviation of {diversity_value:.2f} mm")
```
