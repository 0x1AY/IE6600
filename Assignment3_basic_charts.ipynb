```python
import numpy as np

# Generate 40 random data points between 1.6 and 9.0
np.random.seed(42)  # For reproducibility
data = np.random.uniform(1.6, 9.0, 40)

# Sort the data
sorted_data = np.sort(data)

# Calculate Q1 (25th percentile)
Q1 = np.percentile(sorted_data, 25)

# Calculate Q2 (Median or 50th percentile)
Q2 = np.median(sorted_data)

# Calculate Q3 (75th percentile)
Q3 = np.percentile(sorted_data, 75)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Calculate lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Determine minimum and maximum (excluding outliers)
min_value = sorted_data[sorted_data >= lower_bound][0]
max_value = sorted_data[sorted_data <= upper_bound][-1]

# Identify any outliers
outliers = sorted_data[(sorted_data < lower_bound) | (sorted_data > upper_bound)]

Q1, Q2, Q3, IQR, min_value, max_value, outliers, sorted_data
print(data)
print(sorted_data)
print(f'Q1: {Q1}, Q2: {Q2}, Q3: {Q3}, IQR: {IQR}, MIN: {min_value}, MAX: {max_value}, OUTLIERS: {outliers}')
```

### Question 6

{accuracy: 5}

Wrangle the data to make a new column in the cars dataframe called `Fuel_efficiency` with five categories. The categories should follow from values in the `Miles_per_Gallon` column in the following way:

- If `Miles_per_Gallon` is less than 11.0, the `Fuel_efficiency` should be `Very Low`.
- If `Miles_per_Gallon` is less than 18.0 and greater than or equal to 11.0, the `Fuel_efficiency` should be `Low`.
- If `Miles_per_Gallon` is less than 25.0 and greater than or equal to 18.0, the `Fuel_efficiency` should be `Average`.
- If `Miles_per_Gallon` is less than 30.0 and greater than or equal to 25.0, the `Fuel_efficiency` should be `High`.
- If `Miles_per_Gallon` is greater than or equal to 30.0, the `Fuel_efficiency` should be `Very High`.

Now create a heatmap with `Cylinders` on the x-axis and `Fuel_efficieny` on the y-axis to report `Horsepower`. What general treand do you observe, and how many cylinders would you recommend from the most fuel efficient cars to obtain maximum horsepower.


```python
# YOUR CODE HERE
...
def categorize_fuel_efficiency(mpg):
    if mpg < 11.0:
        return 'Very Low'
    elif 11.0 <= mpg < 18.0:
        return 'Low'
    elif 18.0 <= mpg < 25.0:
        return 'Average'
    elif 25.0 <= mpg < 30.0:
        return 'High'
    elif mpg >= 30.0:
        return 'Very High'

cars['Fuel_efficiency'] = cars['Miles_per_Gallon'].apply(categorize_fuel_efficiency)

# Create a pivot table for the heatmap
pivot_table = cars.pivot_table(values='Horsepower', index='Fuel_efficiency', columns='Cylinders', aggfunc='mean')

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap='coolwarm', cbar_kws={'label': 'Horsepower'})
plt.title('Heatmap of Horsepower by Cylinders and Fuel Efficiency')
plt.xlabel('Cylinders')
plt.ylabel('Fuel Efficiency')
plt.show()

print("""
To obtain maximum horsepower while maintaining the most fuel-efficient cars, it is recommended to choose cars with 4 or 6 cylinders. 
      These cars strike a balance between fuel efficiency and horsepower, providing a good compromise for performance and economy.
""")
```
