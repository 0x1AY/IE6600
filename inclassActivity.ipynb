```python
# Include libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df_superstore = pd.read_csv('https://raw.githubusercontent.com/qurat-azim/instructionaldatasets/main/data/superstore.csv')
df_superstore.head()
```

### In Class activity (~10 min)

Make a pie chart with categories of furniture that also shows percentage of sales as labels


```python
df_furniture = df_superstore[df_superstore['Category'] == 'Furniture']
df_furniture_subcategory = df_furniture.groupby('Sub-Category')['Sales'].sum()
df_furniture_subcategory.plot.pie( y ='Sales', autopct='%1.1f%%', title='Sales by Category of furniture')
```

- Make a heatmap with `Category` on x axis and `Segment` on y azis for `Profit`
- Make a scatterplot with `Discount` on x-axis and `Sales` on y-axis, Colour by `Segment`
- Make a grouped bar chart of `Sales` of different `Segment` options coloured by `Region`


```python
#- Make a heatmap with `Category` on x axis and `Segment` on y azis for `Profit`

df_heatmap = df_superstore.groupby(['Category','Segment'], as_index=False)[['Profit']].sum()
df_heatmap = df_heatmap.pivot(index='Segment', columns='Category', values='Profit')
sns.heatmap(df_heatmap)
```

```python
#- Make a scatterplot with `Discount` on x-axis and `Sales` on y-axis, Colour by `Segment`

sns.scatterplot(data=df_superstore, x='Discount', y='Sales', hue='Segment')
```

```python
#- Make a grouped bar chart of `Sales` of different `Segment` options coloured by `Region`  
df_grouped_bar = df_superstore.groupby(['Segment','Region'], as_index=False)[['Sales']].sum()
df_grouped_bar.head()
sns.barplot(data=df_grouped_bar, x='Sales', y='Segment', hue='Region')
```
