# Basic Data Visualizations


## Imports


```python
# Include libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Reading the data file through pandas


```python
df_superstore = pd.read_csv('https://raw.githubusercontent.com/qurat-azim/instructionaldatasets/main/data/superstore.csv')
df_superstore
```

```python
df_superstore.info()
```

```python
df_superstore['Category'].unique()
```

```python
df_superstore['Order Date'] = pd.to_datetime(df_superstore['Order Date'])
df_superstore.info()
```

```python
df_superstore['Category'].unique()
```

### Bar Chart


```python
df1 = df_superstore[['Category', 'Sales','Profit']]
df1.head()
```

```python
df2 = df1.groupby('Category', as_index=False).sum()
df2
```

```python
df_temp = df_superstore.groupby('Category', as_index=False)[['Sales', 'Profit']].sum()
df_temp
```

```python
sns.barplot(data = df_temp, x = 'Category', y = 'Sales');
```

```python
sns.barplot(data = df_temp, x = 'Category', y = 'Profit');
```

### Pie Chart


```python
df_temp = df_superstore.groupby('Category')[['Sales', 'Profit']].sum() # uses index as labels
df_temp.plot.pie(y='Sales',legend=False)
```

```python

explode = (0.2, 0, 0)  # explode 1st slice
df_temp.plot.pie(y='Profit',legend=False, autopct='%1.1f%%', explode=explode, shadow=True, startangle=90, xlabel='Category')
```

### Line plot


```python
df_temp = df_superstore.groupby('Category',as_index=False)[['Sales', 'Profit']].sum()
df_temp
```

```python
sns.lineplot(x = 'Category', y = 'Sales', data = df_temp)
```

```python
fig, ax = plt.subplots()
sns.lineplot(x = 'Category', y = 'Sales', data = df_temp,ax = ax)
ax.set_ylim(0, None)
plt.show()
```

### Scatter plot


```python
sns.scatterplot(x = 'Sales', y='Profit', data=df_superstore)
```

### Heatmap


```python
df_temp = df_superstore.groupby(['Category','Region'],as_index=False)['Sales'].sum()
df_temp
```

```python
df_temp = df_temp.pivot(index = "Category",
                        columns = "Region",
                        values = "Sales")
df_temp
```

```python
sns.heatmap(df_temp)
```

## Multivariate Encoding

### Grouped Bar Chart


```python
df_temp = df_superstore.groupby(['Category','Region'],as_index=False)[['Sales', 'Profit']].sum()
sns.barplot(y = 'Category', x = 'Profit', hue = 'Region', dodge=True, data = df_temp)
```

### Scatterplot with color encoding


```python
sns.scatterplot(x = 'Sales', y='Profit', hue='Category', data=df_superstore)
```

```python
sns.scatterplot(y = 'Profit', x = 'Sales', size='Quantity', data = df_superstore, hue='Quantity')

```

```python
sns.scatterplot(y = 'Profit', x = 'Sales', size='Quantity', hue='Category', data = df_superstore)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

```python
sns.scatterplot(y = 'Profit', x = 'Sales', size='Quantity', hue='Category', style="Region",data = df_superstore)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

```python
df_temp = df_superstore.groupby(['Category','Region'],as_index=False)[['Sales', 'Profit']].sum()
sns.lineplot(y = 'Profit', x = 'Region', style='Category', hue='Category', errorbar=None, data = df_temp)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```
